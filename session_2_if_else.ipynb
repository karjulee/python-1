#### If magnesium is more than 10 and less than 20, and chloride is more than 5, show ALERT; otherwise show OK.

```python
Mg = 15
Cl = 5.1
if Mg > 10 and Mg < 20 and Cl > 5:
    print (f'ALERT')
else:
    print (f'its okay')
```

In groundwater testing, if calcium (Ca²⁺) level is more than 40 and less than 100 mg/L and nitrate (NO₃⁻) level is more than 10 mg/L, the water may be harmful.

```python
Ca = 45
NO = 12
if Ca > 40 and Ca < 100 and NO > 10:
    print(f'harmful')
```
