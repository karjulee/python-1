```python
import pandas as pd
```

```python
data_one = {'A': ["A1", "A2", "A3"],
            'B': ["B1", "B2", "B3"],
}
data_two = { 'C': ["C1", "C2", "C3"],
             'D': ["D1", "D2", "D3"],
}
```

```python
df_one = pd.DataFrame(data_one)
df_two = pd.DataFrame(data_two)
```

```python
df_one
```

```python
df_two
```

```python
abcd = pd.concat([df_one, df_two], axis=0)
abcd
```

```python
abcd2 = pd.concat([df_one.T, df_two.T], axis=0)
abcd2
```

```python
## 
registration = pd.DataFrame({
    'StudentID': [4, 2, 3, 1, 5],
    'Course': ['Math', 'Science', 'History', 'English', 'Music']
})

Fee_status = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Fee_Paid': [True, False, True, False]
})
```

```python
registration
```

```python
Fee_status
```

```python
pd.concat([registration, Fee_status], axis=1)
```

```python
registration.merge(Fee_status, on='StudentID', how='left') # left means keep all records from left table
```

```python
registration.merge(Fee_status, on='StudentID', how='right') # right means keep all records from right table
```

```python
registration.merge(Fee_status, on='StudentID', how='inner') # inner means keep only matching records
```

```python
registration.merge(Fee_status, on='StudentID', how='outer') # outer means keep all records
```
