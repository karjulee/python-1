### Strength of Material Equation (Based on Temperature and Pressure)

\[
$\sigma$ = 400 (1 - 0.002(T - 25)) + 0.05(P - 1)
\]

#### Where:
- **σ** — Strength of material (MPa)  
- **T** — Temperature (°C)  
- **P** — Pressure (MPa)  



```python
#temp = -10 to 10 with step size of 0.01
#P = 0 to 20 0.1
```

```python
import matplotlib.pyplot as plt
```

```python
T = np.arange(-5,5,0.01)
Pressure = [-5, 0,5,10,15]
for P in Pressure:
    print("pressure",P)
    sigma = 400*(1 - 0.002*(np.log(T) - 25)) + 0.05*(np.exp(-3*P) - 1)
    print(sigma)
    plt.plot(T,sigma,label=str(P))
plt.legend()
    
```
