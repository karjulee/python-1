```python
import pdfplumber

text = "" 
with pdfplumber.open("paper.pdf") as pdf:
    for page in pdf.pages:
        text += page.extract_text()
```

```python
print(text)
```

```python
import ollama


response = ollama.generate("deepseek-r1:8b",
    prompt=f"Summarize the following research paper:\n\n{text}\n\nSummary:",

)
```
