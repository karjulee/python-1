```python
from kivy.app import App
from kivy.uix.widget import Widget
```

```python
# For now we have just an empty window
import kivy.uix.label as Label
class Task01(Widget):
    def __init__(self,**keywords):
        super().__init__(self,**keywords)
        self.add_widget(Label(text="hello karju"))
    


# Just a single empty window
class MyApp(App):
    def build(self):
        return Task01()


#
MyApp().run()
```
