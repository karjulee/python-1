```python
class Cat:
    def __init__(self,name,color):
        self.name = name
        self.color = color
    def describe_it(self):
        print(f"the name of cat is {self.name} and color is {self.color}")
        
```

```python
c1 = Cat("miao",'red')
c2 = Cat("lela","pink")
```

```python
c1.describe_it()
```

```python
c2.describe_it()
```

```python
class Lion:
    def __init__(self,name,color,status="sleeping"):
        self.name = name
        self.color = color
        self.status = status
    def describe_it(self):
        print(f"the name of Lion is {self.name} and color is {self.color}, is {self.status}")
    def change_status(self,new):
        self.status = new
```

```python
L = Lion("tiger","brown")
L.describe_it()
L.change_status("running")
L.describe_it()
```

```python
# make a class Substance
# it should take a dictionary of items for exampel {"H":2,"O":1}
# it shold have a function to print in this substance you have 2 H and 1 0
```

```python
class Sub:
    def __init__(self,sub):
        self.sub = sub
    def tell_details(self):
        for a,b in self.sub.items():
            print(f"for {a} you have {b} elements")
    def total_elements(self):
        print(sum(self.sub.values()))   
    def add_element(self,name,number):
        self.sub[name] = number
        
```

```python
s1 = Sub({"H":2,"O":1})
s1.tell_details()
```

```python
s2 = Sub({"Na":1,"Cl":1})
s2.tell_details()
```

```python
s1.total_elements()
```

```python
s2.total_elements()
```
