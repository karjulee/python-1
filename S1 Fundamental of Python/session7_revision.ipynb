```python
city = []
pos_x = []
pos_y = []
for jaz in jaziras:
    for name,jaz_cat in jaz.items():
        for name2, jaz_cat2 in jaz_cat.items():
            for details in jaz_cat2:
                for k,v in details.items():
                    if k=='city':
                        city.append(v)
                    if k =='coordinates':
                        pos_x.append(v[0])
                        pos_y.append(v[1])
```
