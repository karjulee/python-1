```python
# make a chemical reaction function
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import rdChemReactions
def chemical_reaction(reactant1,reactant2):
    reaction_smarts = {
        ('H2','O2'):"[H][H].[O][O]>>[H][O][H]",
        ('Na','Cl2'):"[Na].[Cl][Cl]>>[Na][Cl]",
        ('C','O2'):"[C].[O][O]>>[C](=O)=O"
    }
    smarts = reaction_smarts.get((reactant1,reactant2),None)
    if smarts is None:
        return "No Reaction Found"
    rxn = rdChemReactions.ReactionFromSmarts(smarts)
    reactants = [Chem.MolFromSmiles(reactant1), Chem.MolFromSmiles(reactant2)]
    products = rxn.RunReactants(reactants)
    product_smiles = [Chem.MolToSmiles(prod[0]) for prod in products]
    return product_smiles if product_smiles else "No Products Formed"
```

```python
chemical_reaction("O=O","[H][H]")
```
